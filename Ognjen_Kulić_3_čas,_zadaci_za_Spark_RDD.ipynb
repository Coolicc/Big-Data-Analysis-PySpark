####Zadatak 1
Učitati i raspakovati fajl **ml-latest-small.zip** sa adrese https://grouplens.org/datasets/movielens/ 
Fajl ratings.csv uploadovati na vaš Google drive.Mountovati drive u notebooku kako bi fajl bio vidljiv i putanju do fajla dobijate sa desnim klikom copy path
Svaki red u fajlu predstavlja userId,movieId,rating,timestamp, odnosno id korisnika (userId) koji je filmu (dat je njegov movieId) dao ocenu označenu vrednošću rating. Timestamp je datum ocenjivanja u milisekundama 


Inicijalizovati Spark Context

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
sc = spark.sparkContext
```

Ucitati fajl ratings.csv

```
ratingRDD = sc.textFile('/content/drive/My Drive/ratings.csv')
```

Prikazati prvih 5 elemenata

```
ratingRDD.take(5)
```

Pri učitavanju fajla u RDD strukturu jedan element je jedan red u fajlu pa o tome vodite računa. Takođe biće učitano i zaglavlje koje treba preskočiti u obradi. Pitanje na koje treba da odgvorite glasi: Koliko filmova je dobilo ocenu 5?

```
res_rdd = ratingRDD.map(lambda x: x.split(",")).filter(lambda x: x[0]!='userId').map(lambda x: [int(x[0]), int(x[1]), float(x[2]), x[3]]).filter(lambda x: x[2] == 5.0)
res_rdd.count()
```

Zadatak 2

Dodati fajl movies.csv na vaš Google drive iz foldera ml-latest-small koji ste skinuli u prethodnom zadatku. Za svaku godinu prikazite broj filmova po toj godini. Rezultat treba da u obliku godina, broj filmova. Sortirati u opadajućem redosledu po broju filmova

```
moviesRDD = sc.textFile("/content/drive/My Drive/movies.csv")
print('broj linija u fajlu',moviesRDD.count())
#fajl sadrzi i neke linije koje nisu obradive pa ih treba izbaciti
rdd4 = moviesRDD.filter(lambda x:not x.startswith('movieId')).map(lambda x:x.split(',')).filter(lambda x: len(x)==3).filter(lambda x: '(' in x[1])
rdd5 = rdd4.map(lambda x: (x[1][x[1].rfind('(')+1:x[1].rfind(')')]))
print('broj linija u fajlu nakon filtriranja',rdd5.count())
rdd5 = rdd5.sortBy(lambda x:x,ascending=False)

rdd5.countByValue()
```

Zadatak 3

Na osnovu fajlova ratings.csv i movies.csv za svaku godinu odrediti koji je najbolje ocenjen film. Za konacnu ocenu filma uzeti njegovu prosecnu ocenu. Rezultat treba da bude godina, naziv filma, rejting. Sortirati po godini.

```
def maxValue(x,y):
  if (x[1]>=y[1]):
    return x
  else:
    return y
rdd6 = ratingRDD.filter(lambda x: not('rating' in x)).map(lambda x:x.split(',')).map(lambda x:(x[1],(float(x[2]),1)))
prosekPoFilmu = rdd6.reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1])).map(lambda x:(x[0],(x[1][0]/x[1][1])))
filmGodina = rdd4.map(lambda x: (x[0],(x[1][:x[1].rfind('(')-1],x[1][x[1].rfind('(')+1:x[1].rfind(')')])))
joinRDD = filmGodina.join(prosekPoFilmu).map(lambda x: (x[1][0][1],(x[1][0][0],x[1][1])))
joinRDD.reduceByKey(lambda x,y: maxValue(x,y)).sortByKey().map(lambda x: (x[0],x[1][0],x[1][1])).collect()
```

Zadatak 4

Učitati fajlove movies.csv i ratings.csv i ispisati imena filmova koji su najviše puta ocenjivani kao i broj ocena. Za rešavanje zadatka koristiti broadcast metod za smestanje imena svih filmova koja će se na kraju prikazati umesto id filma.

```
movieNames = moviesRDD.filter(lambda x:not x.startswith('movieId')).map(lambda x:x.split(',')).map(lambda x: (x[0], x[1])).collect()
movieNamesBC = sc.broadcast(movieNames)
mostRatRDD = ratingRDD.map(lambda x: x.split(",")).filter(lambda x: x[0]!='userId').map(lambda x: (int(x[1]), 1)).reduceByKey(lambda x,y:x+y).sortBy(lambda x: x[1], ascending=False)
mostRatRDD.map(lambda x: (movieNamesBC.value[x[0]][1],x[1])).take(10)
```

Zadatak 6

Učitati fajl 1800.csv (https://drive.google.com/file/d/1eUcapDZ0zmMcfTBWGUUO-yC1vbpqMiE_/view?usp=sharing) koji opisuje temperature koje su merene određenog dana na određenoj stanici. Filtrirati fajl tako da ostanu samo podaci koji imaju oznaku TMIN. Za svaku stanicu odrediti minimalnu temperaturu koja je izmerena

```
def minValue(x,y):
  if (x[0]<y[0]):
    return x
  else:
    return y
tempMinRDD = sc.textFile('/content/drive/My Drive/1800.csv').map(lambda x: x.split(",")).filter(lambda x: x[2] == 'TMIN')
tempMinRDD2 = tempMinRDD.map(lambda x: (x[0], (x[3]))).reduceByKey(lambda x,y: minValue(x,y))
tempMinRDD2.collect()
```

Zadatak 7

Učitati fajl Book.txt (https://drive.google.com/open?id=1-VYU1vQs5W-aj7FDSIcOUliyqQUpDIKq) i prebrojati koliko puta se svaka reč pojavljuje.Ignorišite velika i mala slova. Izbacite znake interpunkcije (,.!,?). Sortirati tako da se na vrhu nađu reči koje se najviše pojavljuju

```
import string
bookRDD = sc.textFile('/content/drive/My Drive/Book.txt').flatMap(lambda x: x.split(' ')).filter(lambda x: x!='').map(lambda x: x.lower())
bookRDD2 = bookRDD.map(lambda x: x.translate(str.maketrans('', '', '!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~'))).map(lambda x: (x,1)).reduceByKey(lambda x,y: x+y)
bookRDD3 = bookRDD2.sortBy(lambda x: x[1], ascending=False)
bookRDD3.collect()
```
